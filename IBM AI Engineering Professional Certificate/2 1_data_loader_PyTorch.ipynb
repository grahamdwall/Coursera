<h2 id="data_class">Create a Dataset Class</h2>


In this section, we will use the previous code to build a dataset class. 



Complete the code to build a Dataset class <code>dataset</code>. As before, make sure the even samples are positive, and the odd samples are negative.  If the parameter <code>train</code> is set to <code>True</code>, use the first 30 000  samples as training data; otherwise, the remaining samples will be used as validation data.  


```python
class ConcreteCrackDataset(Dataset):
    # Constructor
    def __init__(self, transform=None, train=True):
        directory = "/resources/data"
        positive = "Positive"
        negative = "Negative"

        # Load positive and negative image file paths
        positive_file_path = os.path.join(directory, positive)
        negative_file_path = os.path.join(directory, negative)

        positive_files = sorted([os.path.join(positive_file_path, file) for file in os.listdir(positive_file_path) if file.endswith(".jpg")])
        negative_files = sorted([os.path.join(negative_file_path, file) for file in os.listdir(negative_file_path) if file.endswith(".jpg")])

        # Ensure equal number of samples in positive and negative sets
        number_of_samples = min(len(positive_files), len(negative_files)) * 2  

        # Interleave positive and negative images
        all_files = [None] * number_of_samples
        all_files[::2] = positive_files[:number_of_samples // 2]  # Even indices → Positive
        all_files[1::2] = negative_files[:number_of_samples // 2]  # Odd indices → Negative

        # Labels: 1 for positive, 0 for negative
        Y = torch.zeros([number_of_samples], dtype=torch.long)
        Y[::2] = 1
        Y[1::2] = 0

        # Train/Validation split
        if train:
            self.all_files = all_files[:30000]  # First 30,000 for training
            self.Y = Y[:30000]
        else:
            self.all_files = all_files[30000:]  # Remaining for validation
            self.Y = Y[30000:]

        self.len = len(self.all_files)  # Dataset length
        self.transform = transform  # Image transformation

    # Get dataset length
    def __len__(self):
        return self.len

    # Get an item (image and label)
    def __getitem__(self, idx):
        image = Image.open(self.all_files[idx])
        y = idx
        plt.imshow(image)
        # Apply transformation if specified
        if self.transform:
            image = self.transform(image)

        return image, y
```
